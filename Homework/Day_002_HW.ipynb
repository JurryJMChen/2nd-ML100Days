```python
import os
import numpy as np
import pandas as pd
```

```python
# 設定 data_path
dir_data = './data/'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
```

### 如果沒有想法，可以先嘗試找出剛剛例子中提到的問題的答案
#### 資料的 row 數以及 column 數

```python
print('Number of rows:', len(app_train))
print('Number of columns:', len(app_train.columns))
```

#### 列出所有欄位

```python
for eachField in app_train.columns:
    print(eachField)
```

#### 截取部分資料

```python
print(app_train[app_train['AMT_INCOME_TOTAL']>150000])
```
